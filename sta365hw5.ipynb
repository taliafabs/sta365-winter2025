```python
import pymc as pm
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import arviz as az
```

```python
# Generate the synthetic data
np.random.seed(123)
num_samples = 100
true_theta = 0
true_tau = 1
data = np.random.normal(loc=true_theta, scale=1/np.sqrt(true_tau), size = num_samples)
# x = stats.norm(loc=0, scale=1).rvs(size=100)

# Plot a histogram
plt.hist(data, bins=20, alpha=0.5, color='lavender', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Observed data histogram")
plt.show()
```

####  2.1
 1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
# Part 1

model1 = pm.Model()

with model1:
    # theta normal prior
    theta = pm.Normal('theta', mu = 0, sigma = 1)
    # tau gamma prior
    tau = pm.Gamma('tau', alpha = 0.5, beta = 0.5)
    x_obs = pm.Normal('x_obs',
                      mu = theta,
                      sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau)),
                      observed = data)
    trace1 = pm.sample(draws=2000, chains=2, tune=100)

az.plot_trace(trace1)
plt.tight_layout()
plt.show()
# pm.model_to_graphviz(model1)
```

```python
# Inference and diagnostics with az.summary
az.summary(trace1, round_to=5)
```

#### 2.2
2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
# Part 2
model2 = pm.Model()

with model2:
  # non-normal theta prior
  theta = pm.Uniform('theta')
  # non-gamma tau prior
  tau = pm.Beta('tau', alpha=1, beta=1)
  x_obs = pm.Normal('x_obs',
                    mu = theta,
                    sigma = pm.Deterministic('sigma', 1 / pm.math.sqrt(tau)),
                    observed = data)
  trace2 = pm.sample(draws=2000, target_accept=0.95)

az.plot_trace(trace2)
plt.tight_layout()
plt.show()
```

```python
az.summary(trace2, round_to=5)
```

#### 2.3

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
# Part 3
model3 = pm.Model()
data = np.random.normal(2, 3, 100)

with model3:
  # another non-normal theta prior
  theta = pm.Laplace('theta', mu=0, b=1)
  # another non-gamma tau prior
  tau = pm.Exponential('tau', lam=1)
  x_obs = pm.Normal('x_obs',
                         mu = theta,
                         sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau)),
                         observed=data)
  trace3 = pm.sample(draws=2000, target_accept=0.95)

az.plot_trace(trace3)
plt.tight_layout()
plt.show()
```

```python
az.summary(trace3, round_to=5)
```

#### Comparing autocorrelations

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

az.plot_autocorr(trace1, var_names = ["theta"], ax=axes[0])
az.plot_autocorr(trace2, var_names = ["theta"], ax=axes[1])
az.plot_autocorr(trace3, var_names = ["theta"], ax=axes[2])
axes[0].set_title("Autocorrelations for 2.1")
axes[1].set_title("Autocorrelations for 2.2")
axes[2].set_title("Autocorrelations for 2.3")
plt.tight_layout()
plt.show()
```
