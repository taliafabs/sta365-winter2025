####  2.1 Normal theta prior, Gamma tau prior
 1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

````python
# Part 1

model1 = pm.Model()

with model1:
    # theta normal prior
    theta = pm.Normal('theta', mu = 0, sigma = 1)
    # tau gamma prior
    tau = pm.Gamma('tau', alpha = 0.5, beta = 0.5)
    x_obs = pm.Normal('x_obs',
                      mu = theta,
                      sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau)),
                      observed = data)
    trace1 = pm.sample(draws=2000, chains=2, tune=100)

az.plot_trace(trace1)
plt.tight_layout()
plt.show()
# pm.model_to_graphviz(model1)
````

````python
# Bayesian inference & diagnostics with az.summary
az.summary(trace1, round_to=5)
````

````python
az.plot_forest(trace1,
               var_names=["theta", "tau"],
               combined=True,
               hdi_prob=0.95,  # 95% high density interval
               r_hat=True)
plt.tight_layout()
plt.show()
````

````python
az.plot_energy(trace1)
````

````python
# Part 2
model2 = pm.Model()

with model2:
  # non-normal theta prior
  theta = pm.Uniform('theta', lower=0, upper=1)
  # non-gamma tau prior
  tau = pm.Beta('tau', alpha=1, beta=1)
  x_obs = pm.Normal('x_obs',
                    mu = theta,
                    sigma = pm.Deterministic('sigma', 1 / pm.math.sqrt(tau)),
                    observed = data)
  trace2 = pm.sample(draws=2000, target_accept=0.95)

az.plot_trace(trace2)
plt.tight_layout()
plt.show()
````

In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

**How the Markov slice sampling algorithm below works**

Slice Sampling is a technique for generating samples from a probability distribution when the full density is not available, but its unnormalized version can be computed. It works by sampling uniformly from the "slice" (the region where target function $f(x) < y$) beneath the curve defined by the target density function.

The ```slice_f_at_y``` function determines the bounds of the slice, adjusts the interval bounds slightly to ensure the endpoints  $f(a)> y$ and $f(b)>y$ are in the slice, draws a sample $x_$ from that interval, checks if x_ is under the slice ($f(x_) > y$), and accepts $x_ if it is.
If $x_$ is outside the slice, ```slice_f_at_y``` checks whether it was above on the left, or above on the right and adjusts the interval to continue in the recursive case with a reduced interval.

The ```slice_f_at_y_``` function recursively adjusts the search for a sample in the slice, and if the sample $x_$ is outside the slice, it adjuts the interval and continues.



**Slice sampling used in place of Metropolis-Hastings (MH) Step in a Metropolis-within-Gibbs (MwG) sampler**
A Metropolis-within-Gibbs sampler uses Metropolis Hastings to ...
In Metropolis, we really only know the sampling distribution up to a normalizing constant. We can use slice sampling in the place of the MH step in a MwG sampler by sampling directly from the unnormalized full conditional distribution. We would start with the current value of x, use the current value of the normalized density to determine slice height $y$, sample from the slice, (the curve that we would be sampling beneath is the region where $f(x) > y$), and update the chain by accepting the new variable and then continue to the next iteration of the Gibbs sampler.

````python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
````

````python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
````
